## Setup

```python
import os
import pandas as pd
```

## Images, Directories and Labels

```python
labels_2020 = pd.read_csv("/mnt/d/skin-lesion-data/2020/ISIC_2020_Training_GroundTruth.csv")
labels_2020.head()
```

```python
labels_2020['image_name'] = labels_2020['image_name'].apply(lambda name: name+"_2020.jpg")
```

```python
labels_2020 = labels_2020[labels_2020['diagnosis']!='unknown']
```

```python
labels_2020 = labels_2020[~labels_2020['diagnosis'].isin(["cafe-au-lait macule","atypical melanocytic proliferation"])]
```

```python
labels_2020
```

## Renaming the files

```python
for img in os.listdir("/mnt/d/skin-lesion-data/2020/ISIC_2020_Training_JPEG/train/"):
    dst_name = img.split(".")[0]+"_2020.jpg"
    os.rename(src=f"/mnt/d/skin-lesion-data/2020/ISIC_2020_Training_JPEG/train/{img}",
              dst=f"/mnt/d/skin-lesion-data/2020/ISIC_2020_Training_JPEG/train/{dst_name}")
```

```python
labels_2020['diagnosis'].unique()
```

## Mapping

```python
mapping_2020 = {"nevus": "NV", "melanoma": "MEL", "seborrheic keratosis": "BKL",
                "lichenoid keratosis": "BKL", "solar lentigo": "BKL", "lentigo NOS": "BKL"}
```

```python
labels_2020['diagnosis'] = labels_2020['diagnosis'].replace(mapping_2020)
```

```python
labels_2020 = labels_2020['']
```

```python
for img in os.listdir('/mnt/d/skin-lesion-data/2020/ISIC_2020_Training_JPEG/train/'):
    if img in labels_2020['image_name'].values:
        print(f"{img} in directory")
    else:
        os.remove(f'/mnt/d/skin-lesion-data/2020/ISIC_2020_Training_JPEG/train/{img}')
```

```python
labels_2020.to_csv("/mnt/d/skin-lesion-data/2020/ISIC_2020_Training_JPEG/updated_2020_labels.csv", index=False)
```
