```python
import os
import boto3
from pathlib import Path
from tqdm import tqdm
```

```python
BUCKET_NAME = "skin-lesion-data-bucket"
FOLDER_PATH = "/mnt/d/skin-lesion-data/final-data/train/"
S3_PREFIX = "train"
```

```python
IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".gif", ".bmp", ".webp", ".tiff", ".svg"}

def upload_folder_to_s3(folder_path, bucket_name, s3_prefix="train"):
    s3 = boto3.client("s3")

    folder = Path(folder_path)
    files = [f for f in folder.iterdir() if f.is_file() and f.suffix.lower() in IMAGE_EXTENSIONS]

    if not files:
        print("No image files found.")
        return

    for file_path in tqdm(files, desc="Uploading images", unit="file"):
        s3_key = f"{s3_prefix}/{file_path.name}"

        try:
            s3.upload_file(str(file_path), bucket_name, s3_key)
        except Exception as e:
            tqdm.write(f"Failed to upload {file_path.name}: {e}")
```

```python
upload_folder_to_s3(FOLDER_PATH, BUCKET_NAME, S3_PREFIX)
```
