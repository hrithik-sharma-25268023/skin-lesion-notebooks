## Paths

```python
data_2017_2018 = pd.read_csv("/mnt/d/skin-lesion-data/final-data/labels/labels-2017-2018.csv")
data_2019_2020 = pd.read_csv("/mnt/d/skin-lesion-data/final-data/labels/final_combined_labels.csv")
```

```python
data_2017_2018 = data_2017_2018.drop(data_2017_2018.columns[0], axis=1)
```

```python
data_2019_2020
```

```python
labels = pd.concat([data_2019_2020, data_2017_2018], axis=0)
```

```python
# labels.to_csv("/mnt/d/skin-lesion-data/final-data/labels/labels.csv")
```

```python
images = os.listdir("/mnt/d/skin-lesion-data/final-data/train/")
```

```python
mapping = {}

for image in tqdm(images):
    if image in labels['image'].unique():
        mapping[image] = labels[labels['image']==image]['label'].values[0]

# {image: labels[labels['image']==image]['label'].values[0] for image in images if image in labels['image'].unique()}
```

```python
with open('/mnt/d/skin-lesion-data/final-data/labels/mappings.pkl', 'wb') as file:
    pickle.dump(mapping, file)
```
