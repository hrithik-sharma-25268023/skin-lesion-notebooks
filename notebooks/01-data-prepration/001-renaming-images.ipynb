## Introduction

This notebook contains work related to renaming the image files for specific years.

## Setup

```python
import os
from tqdm import tqdm
from skin_lesion_project.utils.data_prepration import rename_files
```

## Images Files

```python
PATH_2018 = "../../data/2018/ISIC2018_Task1-2_Training_Input/ISIC2018_Task1-2_Training_Input/"
PATH_2019 = "../../data/2019/ISIC_2019_Training_Input/ISIC_2019_Training_Input/"
```

```python
images_2018 = [img for img in os.listdir(PATH_2018) if img.endswith("jpg")]
images_2019 = [img for img in os.listdir(PATH_2019) if img.endswith("jpg")]
```

## Renaming Files

### 2018

```python
for img in tqdm(images_2018):
    src = PATH_2018+img
    dst = src.replace('.jpg', '_2018.jpg')
    rename_files(source_path=src, destination_path=dst)
```

### 2019

```python
for img in tqdm(images_2019):
    src = PATH_2019+img
    dst = src.replace('_2018.jpg', '_2019.jpg')
    rename_files(source_path=src, destination_path=dst)
```
