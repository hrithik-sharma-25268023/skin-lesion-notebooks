## Introduction

Combining the 2019 and 2020 images data.

## Setup

```python
import os
import shutil
import pandas as pd
from tqdm import tqdm
```

## Paths

```python
data_2019_dir = "/mnt/d/skin-lesion-data/2019/ISIC_2019_Training_Input/ISIC_2019_Training_Input/"
labels_2019_dir = "/mnt/d/skin-lesion-data/2019/ISIC_2019_Training_GroundTruth_2019.csv"

data_2020_dir = "/mnt/d/skin-lesion-data/2020/ISIC_2020_Training_JPEG/train/"
labels_2020_dir = "/mnt/d/skin-lesion-data/2020/ISIC_2020_Training_JPEG/updated_2020_labels.csv"
```

## Combining the data

### Images

```python
for img in tqdm(os.listdir(data_2019_dir)):
    shutil.copy(src=data_2019_dir+img, dst=f"/mnt/d/skin-lesion-data/2019+2020/train/{img}")
```

```python
for img in tqdm(os.listdir(data_2020_dir)):
    shutil.copy(src=data_2020_dir+img, dst=f"/mnt/d/skin-lesion-data/2019+2020/train/{img}")
```

### Labels

```python
shutil.copy(labels_2019_dir, "/mnt/d/skin-lesion-data/2019+2020/labels/ISIC_2019_Training_GroundTruth_2019.csv")
shutil.copy(labels_2020_dir, "/mnt/d/skin-lesion-data/2019+2020/labels/updated_2020_labels.csv")
```
