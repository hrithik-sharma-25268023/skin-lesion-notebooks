## Introduction

Combining the labels in one file

## Setup

```python
import pandas as pd
from skin_lesion_project.utils import eda_utils
```

```python
class_labels = eda_utils.LABEL_COLS
class_labels
```

```python
class_names = eda_utils.CLASS_NAMES
class_names
```

## Combine

```python
train_2019 = pd.read_csv("/mnt/d/skin-lesion-data/2019/ISIC_2019_Training_GroundTruth_2019.csv")
train_2020 = pd.read_csv("/mnt/d/skin-lesion-data/2020/ISIC_2020_Training_JPEG/updated_2020_labels.csv")

train_2019 = train_2019[["image"] + class_labels]
```

```python
train_2019
```

```python
class_labels
```

```python
labels_2019  = {im: class_labels[int(train_2019.loc[train_2019["image"] == im].drop("image", axis=1).values[0].argmax())] for im in train_2019["image"].values}
labels_2020 = {u:v for u, v in train_2020[['image_name', 'diagnosis']].values}
```

```python
labels = pd.concat([pd.DataFrame({'image':labels_2019.keys(), 'label':labels_2019.values()}), pd.DataFrame({'image':labels_2020.keys(), 'label':labels_2020.values()})], axis=0)
```

```python
labels.to_csv("/mnt/d/skin-lesion-data/2019+2020/labels/final_combined_labels.csv", index=False)
```
